```python
import torch
import torchvision
import matplotlib.pyplot as plt
import numpy as np
import utils
import models
from sklearn.svm import SVC, LinearSVC
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.decomposition import PCA, KernelPCA
from torchsummary import summary
import matplotlib.patches as mpatches
from matplotlib.cm import get_cmap
```

```python
train_dataset = utils.dataset_with_indices(torchvision.datasets.FashionMNIST)(root='mnist', train=True, transform=torchvision.transforms.ToTensor())
test_dataset = utils.dataset_with_indices(torchvision.datasets.FashionMNIST)(root='mnist', train=False, transform=torchvision.transforms.ToTensor())
model = models.Network()
```

```python
train_losses, val_accuracies_regular, val_accuracies_forward, training_time = utils.train(model, train_dataset, test_dataset, utils.supervised_triplet_loss, epochs=20, batch_size=100)
```

```python
train_losses2, val_accuracies_regular2, val_accuracies_forward2, training_time2 = utils.train(model, train_dataset, test_dataset, utils.supervised_triplet_loss, epochs=20, batch_size=100)
```

```python
train_losses3, val_accuracies_regular3, val_accuracies_forward3, training_time3 = utils.train(model, train_dataset, test_dataset, utils.supervised_triplet_loss, epochs=20, batch_size=100)
```

```python
plt.plot(val_accuracies_regular + val_accuracies_regular2 + val_accuracies_regular3, marker='o', label='Test Loss')
plt.xlabel('Iteration Number')
plt.ylabel('Accuracy')
plt.show()
```

```python
all_val_accuracies = val_accuracies_regular + val_accuracies_regular2 + val_accuracies_regular3
np.savetxt('data/triplet_val_accuracies', np.array(all_val_accuracies), delimiter=',')
```

```python
all_train_losses = train_losses + train_losses2 + train_losses3
np.savetxt('data/triplet_train_losses', np.array(all_train_losses), delimiter=',')
```

```python
plt.plot(train_losses + train_losses2 + train_losses3, marker='o', label='Training Loss')
plt.xlabel('Iteration Number')
plt.ylabel('Loss')
plt.show()
```

```python
train_embedding, test_embedding = utils.embed_regular(model, train_dataset, test_dataset)

raw_train_embedding = train_dataset.data.reshape(train_dataset.data.shape[0], 28*28)
raw_test_embedding = test_dataset.data.reshape(test_dataset.data.shape[0], 28*28)
```

```python
svm_train_embedding = LinearSVC(C=10.0).fit(train_embedding, train_dataset.targets)
svm_train_embedding.score(test_embedding, test_dataset.targets)
```

```python
pca_embedding = PCA(n_components=2).fit(train_embedding)
pca_test_projection = pca_embedding.transform(test_embedding)

pca_raw = PCA(n_components=2).fit(raw_train_embedding)
pca_raw_projection = pca_raw.transform(raw_test_embedding)
```

```python
fig, axs = plt.subplots(2, figsize=(6, 10))
ax1 = axs[0]
ax2 = axs[1]

# plot 1
ax1.scatter(pca_raw_projection[:,0], pca_raw_projection[:,1], c=test_dataset.targets, cmap='tab10', s=1)
ax1.set_xlabel('$x_1$')
ax1.set_ylabel('$x_2$')
ax1.set_title('PCA Projection of Raw Data')

# plot 2
ax2.set_title('PCA Projection of Embedded Data')
ax2.scatter(pca_test_projection[:,0], pca_test_projection[:,1], c=test_dataset.targets, cmap='tab10', s=1)
ax2.set_xlabel('$x_1$')
ax2.set_ylabel('$x_2$')

patches = []
num_classes = len(train_dataset.classes)
for i in range(num_classes):
    patches.append(mpatches.Patch(color=get_cmap('tab10')(i/(num_classes-1)), label=train_dataset.classes[i]))
plt.legend(handles=patches, loc='center', bbox_to_anchor=(0.5, -0.25), ncol=5)
plt.show()
```

```python
lda_embedding = LinearDiscriminantAnalysis(n_components=2).fit(train_embedding, train_dataset.targets)
lda_test_projection = lda_embedding.transform(test_embedding)

lda_raw = LinearDiscriminantAnalysis(n_components=2).fit(raw_train_embedding, train_dataset.targets)
lda_raw_projection = lda_raw.transform(raw_test_embedding)
```

```python
fig, axs = plt.subplots(2, figsize=(6, 10))
ax1 = axs[0]
ax2 = axs[1]

# plot 1
ax1.scatter(lda_raw_projection[:,0], lda_raw_projection[:,1], c=test_dataset.targets, cmap='tab10', s=1)
ax1.set_xlabel('$x_1$')
ax1.set_ylabel('$x_2$')
ax1.set_title('LDA Projection of Raw Data')

# plot 2
ax2.set_title('LDA Projection of Embedded Data')
ax2.scatter(lda_test_projection[:,0], lda_test_projection[:,1], c=test_dataset.targets, cmap='tab10', s=1)
ax2.set_xlabel('$x_1$')
ax2.set_ylabel('$x_2$')

patches = []
num_classes = len(train_dataset.classes)
for i in range(num_classes):
    patches.append(mpatches.Patch(color=get_cmap('tab10')(i/(num_classes-1)), label=train_dataset.classes[i]))
plt.legend(handles=patches, loc='center', bbox_to_anchor=(0.5, -0.25), ncol=5)
plt.show()
```
