```python
import torch
import torchvision
import utils
import numpy as np
import matplotlib.pyplot as plt
```

```python
data, labels = utils.read_data('cifar-10-batches-py/data_batch_1')
augmented_data = utils.augment(data)
for i in range(3):
    plt.imshow(data[i].transpose(1, 2, 0))
    plt.show()
    plt.imshow(augmented_data[i].transpose(1, 2, 0))
    plt.show()
    plt.imshow((data[i] - augmented_data[i]).transpose(1, 2, 0))
    plt.show()
```

```python
batch1_file_name = 'cifar-10-batches-py/data_batch_1'
dict = utils.unpickle(batch1_file_name)
labels = np.array(dict[b'labels'])
data = dict[b'data']
```

```python
data = data.reshape(10000, 3, 32, 32)
```

```python
for i in range(3):
    plt.imshow(data[i].transpose(1, 2, 0))
    plt.show()
```

```python
aa = torchvision.transforms.AutoAugment(policy=torchvision.transforms.AutoAugmentPolicy.CIFAR10)
input_tensor = torch.tensor(data, dtype=torch.uint8)
result = aa.forward(input_tensor).numpy()
for i in range(3):
    plt.imshow(result[i].transpose(1, 2, 0))
    plt.show()
```
