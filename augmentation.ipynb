```python
import matplotlib.pyplot as plt
import torch
import torchvision
import utils
import numpy as np
```

```python
train_dataset = utils.dataset_with_indices(torchvision.datasets.FashionMNIST)(root='mnist', train=True, transform=torchvision.transforms.ToTensor())
test_dataset = utils.dataset_with_indices(torchvision.datasets.FashionMNIST)(root='mnist', train=False, transform=torchvision.transforms.ToTensor())
```

```python
image = train_dataset.data[np.random.randint(60000)]
ra1 = torchvision.transforms.RandAugment()
ra2 = torchvision.transforms.RandAugment()
augmented_image1 = ra1(image.unsqueeze(0))
augmented_image2 = ra2(image.unsqueeze(0))
plt.imshow(image, cmap='binary')
plt.show()
plt.imshow(augmented_image1[0], cmap='binary')
plt.show()
plt.imshow(augmented_image2[0], cmap='binary')
plt.show()
```

```python
aa = torchvision.transforms.AutoAugment()
plt.imshow(aa(image.unsqueeze(0))[0], cmap='binary')
plt.show()
```

```python
ra = torchvision.transforms.RandAugment()
plt.imshow(ra(image.unsqueeze(0))[0], cmap='binary')
plt.show()
```

```python
degrees = 15
ca = torch.nn.Sequential(
    torchvision.transforms.RandomHorizontalFlip(),
    torchvision.transforms.AugMix(severity=5)
)
plt.imshow(ca(image.unsqueeze(0))[0], cmap='binary')
plt.show()
```
